## **Ejercicio 1: Población y Muestra**

En este ejercicio vamos a comprender los conceptos de **población** y **muestra** usando Python.  

- **Población**: Conjunto completo de datos que nos interesa analizar (por ejemplo, todas las alturas de una ciudad).  
- **Muestra**: Subconjunto de la población que seleccionamos para hacer inferencias.  

Vamos a simular una población de alturas y luego extraer una muestra aleatoria para compararlas.


---



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Semilla para reproducibilidad
np.random.seed(75)
```

```python
# Generar población: alturas con media 170cm y desviación estándar 10cm
poblacion = np.random.normal(loc=170, scale=10, size=1000)
```

```python
# Extraer muestra aleatoria de tamaño 100
muestra = np.random.choice(poblacion, size=100, replace=False)
```

```python
# Mostrar estadísticas
print(f"Tamaño de la población: {len(poblacion)}")
print(f"Tamaño de la muestra: {len(muestra)}")
print(f"Media población: {poblacion.mean():.2f} cm")
print(f"Media muestra: {muestra.mean():.2f} cm")
```

```python
# Graficar histogramas para comparar
plt.hist(poblacion, bins=30, alpha=0.5, label='Población')
plt.hist(muestra, bins=30, alpha=0.8, label='Muestra')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.legend()
plt.title('Distribución de alturas: población vs muestra')
plt.show()
```
